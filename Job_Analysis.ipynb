### **2. Data Cleaning**

**2.1) Dealing with Null Values**

- **Removed Rows:**
  - Entries with missing values in `job_id`, `resposibilities`, and `company` were completely removed.

- **Imputed Values:**
  - **Experience:** Missing values filled with `5-10 Yrs`
  - **Location:** Missing values filled with `Bangalore/Bengaluru`
  - **Rating:** Missing values filled with `0.0` (indicating no rating)
  - **Reviews:** Missing values filled with `0 Reviews`


```python
df = df.dropna(subset = ['job_id','resposibilities','company'])
```

```python
df['experience'] = df['experience'].fillna('5-10 Yrs')
df['location'] = df['location'].fillna('Bangalore/Bengaluru')
df['rating'] = df['rating'].fillna(0.0)
df['reviews'] = df['reviews'].fillna('0 Reviews')
```

**2.2) Dealing with Duplicates**

- All duplicate entries based on the job_id column have been successfully removed.
- This ensures each job listing is unique and consistent, preventing redundancy in the analysis.

```python
df = df.drop_duplicates(subset = 'job_id')
```

**2.3) Change the Datatype of job_id**

```python
df['job_id'] = df['job_id'].astype('int').astype(str)
```

**2.4) Clean Experience Column**

```python
df['min_experience'] = df['experience'].str.replace('Yrs','').str.split('-').str[0].str.strip().astype('int')
df['max_experience'] = df['experience'].str.replace('Yrs','').str.split('-').str[1].str.strip().astype('int')
```

**2.5) Clean Salary Column**

```python
df['clean_salary'] = df['salary'].str.replace(' PA.','').str.replace(',','').str.split('-').str[0].str.strip()

df = df[(df['clean_salary'] != '9.5 Cr and above') & (df['clean_salary'] != 'Less than 5000') &  (df['clean_salary'] != 'Less than 50000')]
df['min_salary'] = df['clean_salary'].str.split('(').str[0].str.strip()
del df['clean_salary']


df['clean_salary'] = df['salary'].str.replace(' PA.','').str.replace(',','').str.split('-').str[1].str.split('(').str[0].str.strip()
df = df[df['clean_salary'] != '9.5 Cr and above']
df['max_salary'] = df['clean_salary'].fillna('Not disclosed')
del df['clean_salary']

del df['salary']
del df['experience']
```

**2.6) Clean review columns**

```python
df['reviews'] = df['reviews'].str.split(' ').str[0].astype('int')
```

### **3. Data Analysis**

**3.1) Overall Analysis**

```python
print('-'*30)
print('Total Jobs         : ', len(df))
print('Total Companies    : ', df['company'].nunique())
print('-'*30)
```

**3.2) Top 5 Companies(by Reviews)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter


# Group by company and get top 5
top_companies = (
    df.groupby("company")["reviews"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)
top_companies.columns = ["company_name", "reviews"]

# Plotting
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
barplot = sns.barplot(
    x="reviews",
    y="company_name",
    data=top_companies,
    palette="viridis"
)

# Format x-axis with commas
barplot.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))

# Add number labels to bars
for index, row in top_companies.iterrows():
    plt.text(
        row.reviews + 1e6,  # slightly offset to the right of the bar
        index,
        f"{int(row.reviews):,}",
        va='center',
        fontsize=10,
        color='black'
    )

# Set titles and labels
plt.title("Top 5 Companies by Total Reviews", fontsize=16, fontweight='bold')
plt.xlabel("Total Reviews ", fontsize=12, fontweight='bold')
plt.ylabel("Company Name", fontsize=12, fontweight='bold')
plt.tight_layout()

# Show plot
plt.show()
```

**3.4) Skills needed to become a Data Analyst**

```python
import matplotlib.pyplot as plt
from collections import Counter
import matplotlib.cm as cm
import pandas as pd

# Filter for 'Data Analyst' jobs
data_analyst_jobs = df[df['job_role'] == 'Data Analyst']

# Extract and count responsibilities
responsibilities = data_analyst_jobs['resposibilities'].str.lower().str.split(',').explode().str.strip()
resp_counts = responsibilities.value_counts().head(10)

# Create a DataFrame for plotting
resp_df = resp_counts.reset_index()
resp_df.columns = ['responsibility', 'count']

# Normalize counts for colormap
colors = cm.PuBuGn(resp_df['count'] / float(max(resp_df['count'])))

# Explode top 3
explode = [0.1 if i < 3 else 0 for i in range(len(resp_df))]

# Plot
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(7, 7))

wedges, texts, autotexts = plt.pie(
    resp_df['count'],
    labels=resp_df['responsibility'],
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    textprops=dict(color="black", fontsize=10, fontname='monospace'),
    pctdistance=0.85,
    explode=explode
)

# Highlight top 3 responsibilities
top3_indices = resp_df['count'].nlargest(3).index

for i in range(len(resp_df)):
    if i in top3_indices:
        autotexts[i].set_color('white')
        autotexts[i].set_fontsize(12)
    else:
        autotexts[i].set_fontsize(10)

# Title and final touches
plt.title("Top 10 Responsibilities for Data Analyst", fontsize=16, fontname='monospace', fontweight='bold',pad=20)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

**3.5) Top 10 Job Roles**

```python

# Set plot style
plt.style.use("seaborn-v0_8-whitegrid")

# Prepare data
top_roles = df['job_role'].value_counts().head(10)

# Plot
plt.figure(figsize=(12, 6))
bars = plt.barh(top_roles.index, top_roles.values, color="#4BA3C3")

# Add value labels to bars
for i, v in enumerate(top_roles.values):
    plt.text(v + 10, i, f"{v:,}", va='center', fontsize=10, color='black')


# Title and labels
plt.title("Top 10 Job Roles by Frequency", fontsize=16, fontweight='bold')
plt.xlabel("Number of Job Listings", fontsize=12, fontweight='bold')
plt.ylabel("Job Role", fontsize=12, fontweight='bold')

# Reverse y-axis to show highest value on top
plt.gca().invert_yaxis()

# Improve layout
plt.tight_layout()

# Show plot
plt.show()
```

**3.6) Top 10 Hiring Companies**

```python
import matplotlib.pyplot as plt


labels = top_companies.index
sizes  = top_companies.values

# Explode all slices a bit for separation
explode = [0.03] * len(sizes)

# Use a rich colormap
cmap   = plt.get_cmap('tab20c')
colors = cmap(range(0, 20, int(20/len(sizes))))

# figure size
fig, ax = plt.subplots(figsize=(7, 6), dpi=100)

# Draw the pie chart (donut style)
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=None,
    autopct=lambda p: f"{p:.1f}%\n({int(p/100.*sum(sizes))})",
    startangle=140,
    colors=colors,
    explode=explode,
    shadow=True,
    wedgeprops={'width': 0.4, 'edgecolor': 'white', 'linewidth': 1.5},
    textprops={'fontsize': 9, 'fontweight': 'bold'}
)

# Create a donut center (without text)
centre_circle = plt.Circle((0, 0), 0.65, fc='white')
ax.add_artist(centre_circle)

# Legend on the side
ax.legend(
    wedges,
    labels,
    title="Companies",
    loc="center left",
     bbox_to_anchor=(1, 0, 0.5, 1),
    fontsize=9
)

# Title & layout
ax.set_title(
    "Top 10 Hiring Companies",
    pad=20,
    fontsize=14,
    fontweight='bold'
)

plt.tight_layout()
plt.show()
```

**3.7) Top Job Locations**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data
top_locations = df['location'].value_counts().head(10)

# Use a colormap for dynamic coloring
colors = sns.color_palette("plasma", len(top_locations))

# Plot
plt.figure(figsize=(12, 6))
bars = plt.bar(top_locations.index, top_locations.values, color=colors)

# Add number labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 10, f"{int(height):,}",
             ha='center', va='bottom', fontsize=10, fontweight='bold', color='#333')

# Titles and labels
plt.title("Top 10 Job Locations", fontsize=16, fontweight='bold', color='#2a2a2a')
plt.ylabel("Number of Job Listings", fontsize=12, fontweight='bold')
plt.xlabel("Location", fontsize=12, fontweight='bold')

# Improve x-tick readability
plt.xticks(rotation=45, ha='right')

# Layout
plt.tight_layout()
plt.show()

```

**3.8) Skills to get hired in HDFC Bank**

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns

# Step 1: Prepare the Data
hdfc_responsibilities = (
    df[df['company'].str.lower() == 'hdfc bank']['resposibilities']
    .dropna()
    .str.lower()
    .str.split(',')
    .explode()
    .str.strip()
    .value_counts()
    .head(10)
)

# Reset for plotting
hdfc_df = hdfc_responsibilities.reset_index()
hdfc_df.columns = ['responsibility', 'count']

# Step 2: Normalize for color mapping (copper)
colors = cm.copper(hdfc_df['count'] / float(hdfc_df['count'].max()))

# Step 3: Explode top 3 slices
explode = [0.1 if i < 3 else 0 for i in range(len(hdfc_df))]

# Plot with smaller size
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(7, 7))

wedges, texts, autotexts = plt.pie(
    hdfc_df['count'],
    labels=hdfc_df['responsibility'],
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    textprops=dict(color="black", fontsize=10, fontname='monospace'),
    pctdistance=0.85,
    explode=explode
)

# Highlight top 3
top3_indices = hdfc_df['count'].nlargest(3).index

for i in range(len(hdfc_df)):
    if i in top3_indices:
        autotexts[i].set_color('white')
        autotexts[i].set_fontsize(12)
    else:
        autotexts[i].set_fontsize(10)

# Title and layout
plt.title("Top 10 Responsibilities at HDFC Bank", fontsize=16, fontname='monospace', fontweight='bold', pad=20)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
